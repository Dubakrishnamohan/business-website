```
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import Descriptors
Smiles= input("enter your smiles formulla:")
mol=Chem.MolFromSmiles(Smiles)
img=Draw.MolToImage(mol)
mol.SetProp("name","artemisinin")
display(img)
print(mol.GetProp("name"))
# descriptors access
mol_wt=Descriptors.MolWt(mol)
logP=Descriptors.MolLogP(mol)
H_Acceptors=Descriptors.NumHAcceptors(mol)
H_Donors=Descriptors.NumHDonors(mol)
tpsa=Descriptors.TPSA(mol)
#print descriptors
print("mol_wt: ",mol_wt)
print("LogP: ",logP)
print("H_Acceptors: ",H_Acceptors)
print("H_Donors",H_Donors)
print("tpsa",tpsa)

fail_count= 0
if(mol_wt<=500):
    print("this molecule passes the molecular weight lipsinki rule")
else:
    print("mol_wt: fails")
    fail_count+=1
if(logP<=5):
     print("this molecule passes the logP lipsinki rule")
else:
     print("logP: fails")
     fail_count+=1
if(H_Acceptors<=5):
      print("this molecule passes the H_Acceptors lipsinki rule")
else:
      print("H_Acceptors:fails")
      fail_count+=1
if(H_Donors<=10):
      print("this molecule passes the H_Donors lipsinki rule")
else:
     print("H_Donors: fails")
     fail_count+=1
if(tpsa<=200):
    print("this molecule passes the tpsa lipsinki rule")
else:
    print("tpsa:fails")
if(fail_count==0):
  print("this compound is good for oral healthy")
else:
  print("this compound is not recommend for oral health")


```
